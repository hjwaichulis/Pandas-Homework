```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(file_to_load)
```

## Player Count

```python
unique_players = len(purchase_data_df["SN"].unique())
players_dict = {"Total Players" : [unique_players]}
players_df = pd.DataFrame(players_dict)
players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python

#Calculations
unique_items_df= len(purchase_data_df["Item ID"].unique())
avg_price_df = purchase_data_df["Price"].mean()
num_purchases_df = purchase_data_df["Purchase ID"].count()
total_revenue_df = purchase_data_df["Price"].sum()
#summary df
summary_data = {
    "Number of Unique Items": [unique_items_df],
    "Average Price": [avg_price_df],
    "Number of Purchases": [num_purchases_df],
    "Total Revenue": [total_revenue_df]}
#headers
summary_df = pd.DataFrame(summary_data, columns = ["Number of Unique Items","Average Price",
                                                   "Number of Purchases","Total Revenue"])
summary_df.head()

# throwing an error when i add mapping .map("${:.2f}".format) on avg price/tot rev
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# count players by gender
by_gender = purchase_data_df.groupby("Gender")["SN"].nunique()
by_gender_df = pd.DataFrame(by_gender)

# percent by gender
percent_by_gender = (purchase_data_df["Gender"].value_counts()/purchase_data_df["Gender"].count())*100

# Add % of players column
by_gender_df["Percentage of Players"] = percent_by_gender.map("{:.2f}%".format)
gender_demographics = pd.DataFrame (by_gender_df)
gender_demographics_df = gender_demographics.rename(columns={"SN":"Total Count"})
gender_demographics_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
group_by_gender = purchase_data_df.groupby("Gender")
#Calc Purchase Count
purchase_count = group_by_gender["SN"].count()
purchasecount_df = pd.DataFrame(purchase_count)

#Calc Average Purchase Price
avg_price = group_by_gender["Price"].mean()
purchasecount_df["Average Purchase Price"] = round(avg_price,2)
gender_purchase_df = pd.DataFrame(purchasecount_df)


gender_purchase_df2 = gender_purchase_df.rename(columns={"SN":"Purchase Count"})

#Calc Total Purchase Value
purchase_val = gender_purchase_df2["Purchase Count"].multiply(gender_purchase_df2["Average Purchase Price"], fill_value=1)
gender_purchase_df2['Total Purchase Value'] = round(purchase_val,2).map("${:.2f}".format)

#Calc Avg Total Purchase per Person
gender_count = purchase_data_df.groupby("Gender")["SN"].nunique()
gender_total = purchase_data_df.groupby("Gender")["Price"].sum()
Avg_puch_pp = round(gender_total/gender_count,2)
gender_purchase_df2['Avg Total Purchase per Person'] = round(Avg_puch_pp,2).map("${:.2f}".format)


gender_purchase_df2
# add mapping .map("${:.2f}".format) on avg price
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#binning
bins = [0,9,14,19,24,29,34,39, 44]
age_range = ["<10", "10-14","15-19", "20-24", "25-29", "30-34", "35-39", ">40"]
purchase_data_df["Age Range"] = pd.cut(purchase_data_df["Age"], bins, labels=age_range)

#group and calc total and percent
age_group = purchase_data_df.groupby("Age Range")["SN"].count()
prct_by_age = round(purchase_data_df["Age Range"].value_counts()/unique_players,2)*100

#created df
age_demographic_df = pd.DataFrame({"Total Count":age_group, "Percentage of Players":prct_by_age.map("{:.2f}%".format)})
age_demographic_df

#how do i get <10 on top?
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
total_purchase_value = purchase_data_df.groupby("Age Range")["Price"].sum()
avg_purch_per_person = purchase_data_df.groupby("Age Range")["Price"].mean()
avg_purch_per_person.round(2)
purchasing_analysis = pd.DataFrame({"Total Count":age_group, 
                                    "Percentage of Players":prct_by_age.map("{:.2f}%".format),
                                    "Total Purchase Value":total_purchase_value.map("${:.2f}".format), 
                                    "Avg Total Purchase per Person":avg_purch_per_person.round(2).map("${:.2f}".format)})
purchasing_analysis
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
purchase_count_df = purchase_data_df.groupby("SN").count()["Price"].rename("Purchase Count")
average_price_df = purchase_data_df.groupby("SN").mean()["Price"].rename("Average Purchase Price")
total_purchase_value_df= purchase_data_df.groupby("SN").sum()["Price"].rename("Total Purchase Value")

overall_spending = pd.DataFrame({"Purchase Count":purchase_count_df,
                                "Average Purchase Price":average_price_df.map("${:.2f}".format) ,
                                "Total Purchase Value":total_purchase_value_df.map("${:.2f}".format)})
top_spenders = overall_spending.sort_values("Total Purchase Value", ascending=False)
top_spenders.head(10)
#my Top spenders don't match up withe the solution. Lisosia93 as $18.96
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
purchase_count_df = purchase_data_df.groupby(["Item ID", "Item Name"]).count()["Price"].rename("Purchase Count")
Item_Price =  purchase_data_df.groupby(["Item ID", "Item Name"]).mean()["Price"].rename("Item Price")
total_purchase_value =  purchase_data_df.groupby(["Item ID", "Item Name"]).sum()["Price"].rename("Total Purchase Value")

purchased_items = pd.DataFrame({"Purchase Count":purchase_count_df,
                    "Item Price": Item_Price.map("${:.2f}".format),
                   "Total Purchase Value":total_purchase_value.map("${:.2f}".format)})
most_popular_items = purchased_items.sort_values("Purchase Count", ascending=False)
most_popular_items.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order

```python
most_profitable_items = purchased_items.sort_values("Total Purchase Value", ascending=False)
most_profitable_items.head()
# my highest on this one is not matching up. Oathbreaker, Last Hope of the Breaking Storm has $50.76.  but value shows above??
```
